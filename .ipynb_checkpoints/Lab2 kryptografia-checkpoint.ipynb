```python
def to_list(lista):
    lista2=[]
    for letter in lista:
        lista2.append(letter)
    return lista2
def P10(klucz):
    return [klucz[2],klucz[4],klucz[1],klucz[6],klucz[3],klucz[9],klucz[0],klucz[8],klucz[7],klucz[5]]
def PW(klucz):
    return [klucz[1],klucz[5],klucz[2],klucz[0],klucz[3],klucz[7],klucz[4],klucz[6]]
def SL1(cz1,cz2):
    cz1_p=[cz1[i+1] for i in range(4)]
    cz1_p.append(cz1[0])
    cz2_p=[cz2[i+1] for i in range(4)]
    cz2_p.append(cz2[0])
    return cz1_p,cz2_p
def SL2(cz1,cz2):
    cz1_p1=[cz1[i+2] for i in range(3)]
    cz1_p1.append(cz1[0])
    cz1_p1.append(cz1[1])
    cz2_p2=[cz2[i+2] for i in range(3)]
    cz2_p2.append(cz2[0])
    cz2_p2.append(cz2[1])
    return cz1_p1,cz2_p2
def P10w8(klucz):
    return [klucz[5],klucz[2],klucz[6],klucz[3],klucz[7],klucz[4],klucz[9],klucz[8]]
def zamiana(klucz):
    return [klucz[4],klucz[5],klucz[6],klucz[7],klucz[0],klucz[1],klucz[2],klucz[3]]
def P4w8(cz2):
    return [cz2[3],cz2[0],cz2[1],cz2[2],cz2[1],cz2[2],cz2[3],cz2[0]]
def P4(klucz):
    return [klucz[1],klucz[3],klucz[2],klucz[0]]
def XOR(wejscie,klucz_p):
    wejscie= to_list(wejscie)
    klucz_p= to_list(klucz_p)
    dl=len(wejscie)
    wejscie=("".join(wejscie))
    klucz_p=("".join(klucz_p))
    wejscie=(int(wejscie,2))
    klucz_p=(int(klucz_p,2))
    wynik=bin(wejscie^klucz_p)
    wynik=str(wynik)[2:]
    while len(wynik)<dl:
            wynik="0"+wynik
    wynik=to_list(wynik)    
    return wynik
def PO(klucz):
    return [klucz[3],klucz[0],klucz[2],klucz[4],klucz[6],klucz[1],klucz[7],klucz[5]]
def Runda(klucz,tekst):
    f=open("sdes.txt",'r')
    g=open("sdes2.txt",'r')
    sbox1 = eval(f.read())
    sbox2 = eval(g.read())
    f.close()
    g.close()
    tekst_cz1=tekst[:4:]                    # tekst_cz1 to część pierwsza
    tekst_cz2=tekst[4:8:]                   # tekst_cz2 to nie ruszana cz2
    tekst_cz2_2=tekst_cz2                   # tekst_cz2_2 to kopia na której robimy wszystko  
    tekst_cz2_2=P4w8(tekst_cz2_2)                      #na cześć drugą v2 działamy P4w8
    print("Po P4w8 przed xorem", tekst_cz2_2,klucz)
    tekst_cz2_2=XOR(tekst_cz2_2,klucz)        #na kopie drugiej części i klucz rundy 1 XOR   
    print("Po Xorze", tekst_cz2_2)
    t_cz2_22=tekst_cz2_2[:4:]                 #pierwsze 4 bity 
    t_cz2_24=tekst_cz2_2[4:8:]                #drugie 4 bity
    t_cz2_22=sbox1[int(t_cz2_22[0]),int(t_cz2_22[1]),int(t_cz2_22[2]),int(t_cz2_22[3])]
    t_cz2_24=sbox2[int(t_cz2_24[0]),int(t_cz2_24[1]),int(t_cz2_24[2]),int(t_cz2_24[3])]    #na 2 części i działamy Sboxami
    print("Po sboxach", t_cz2_22,t_cz2_24)
    tekst_cz2_2=P4(t_cz2_22+ t_cz2_24)                       #łączymy i permutacja P4
    print("Po P4 przed 2 xorem", tekst_cz2_2)
    tekst_cz2_2 =[str(i) for i in tekst_cz2_2]
    print("po mapowaniu ",tekst_cz2_2,tekst_cz2)
    tekst_cz2_2=XOR(tekst_cz2_2,tekst_cz1)                  # i kolejny XOR z 4 bitami tekstu
    print("po 2 xorze",tekst_cz2_2,tekst_cz2)
    wynik=tekst_cz2_2+tekst_cz2
    return wynik
def klucze_r(klucz):
    klucz2=P10(klucz)               #klucze są ok
    cz1=klucz2[:5:]
    cz2=klucz2[5:10:] 
    cz=SL1(cz1,cz2)
    #print("częsci klucza 1",cz)
    klucz=cz[0]+cz[1] 
    klucz_rund1=P10w8(klucz) 
    cz=SL2(cz[0],cz[1])
    #print("częsci klucza 2",cz)
    klucz_rund2=P10w8(cz[0]+cz[1])
    print("klucze są takie:",klucz_rund1,klucz_rund2)
    return klucz_rund1,klucz_rund2
def szyfrowanie(klucz,tekst):
    klucze=klucze_r(klucz)
    tekst=PW(tekst)
    #print ("klcze to", klucze)
    print("Tekst po PW: ",tekst)   
    wynik=Runda(klucze[0],tekst)
    print("tekst po rundzie pierwszej to",wynik)
    wynik=zamiana(wynik)
    wynik=Runda(klucze[1],wynik)
    print("tekst po rundzie drugiej to",wynik)
    wynik=PO(wynik)
    return wynik
def odszyfrowywanie(klucz,tekst):
    klucze=klucze_r(klucz)
    tekst=PW(tekst)
    wynik=Runda(klucze[1],tekst)
    wynik=zamiana(wynik)
    wynik=Runda(klucze[0],wynik)
    wynik=PO(wynik)
    return wynik
def s_wejscie(wejscie):
    while len(wejscie)<8:
        wejscie=wejscie+"0"
    return wejscie
wejscie="11111000"
wejscie=s_wejscie(wejscie)
klucz_p= "1100000011"
lista=to_list(wejscie)
klucz=to_list(klucz_p)
print("tekst wejściowy",lista)
wynik2=szyfrowanie(klucz,lista)
print("tekst po zaszyfrowaniu",wynik2)
wynik3=odszyfrowywanie(klucz,wynik2)
print("tekst po odszyfrowaniu",wynik3)
print ("gotowe!")
```
