```python
import csv
from matplotlib import use
use('Agg')
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
#wypakowanie danych do listy
with open('titanic.csv', 'r') as f:
    reader = csv.reader(f)
    lista = list(reader)
lista_3M=0
lista_3W=0
lista_1M=0
lista_1W=0
lista_3M_s=0
lista_3W_s=0
lista_1M_s=0
lista_1W_s=0
for i in range(len(lista)):
    if lista[i][4]=="female":
        if lista[i][2]=="3" and lista[i][1]=="1":
            lista_3W_s+=1 
        if lista[i][2]=="3" and lista[i][1]=="0":
            lista_3W+=1
        if lista[i][2]=="1" and lista[i][1]=="0":
            lista_1W+=1
        if lista[i][2]=="1" and lista[i][1]=="1":
            lista_1W_s+=1
    elif lista[i][4]=="male":
        if lista[i][2]=="1" and lista[i][1]=="0":
            lista_1M+=1
        if lista[i][2]=="1" and lista[i][1]=="1":
            lista_1M_s+=1
        elif lista[i][2]=="3" and lista[i][1]=="0":
            lista_3M+=1
        elif lista[i][2]=="3" and lista[i][1]=="1":
            lista_3M_s+=1
print( lista_3M,lista_3W,lista_1M,lista_1W,lista_3M_s,lista_3W_s,lista_1M_s,lista_1W_s)



n_groups = 2
men = (lista_3M, lista_3M_s)
women = (lista_3W, lista_3W_s)

(fig,ax) = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.30

opacity = 0.4
error_config = {'ecolor': '0.3'}
rects1 = plt.bar(index + bar_width, men, bar_width,
                 alpha=opacity,
                 color='b',
                 error_kw=error_config,
                 label='male')

rects2 = plt.bar(index, women, bar_width,
                 alpha=opacity,
                 color='r',
                 error_kw=error_config,
                 label='female')

ax.set_xlim(-bar_width,len(index)+bar_width)
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival rate 3rd class')
plt.xticks(index + bar_width, ('No', 'Yes'))
plt.legend()

plt.tight_layout()
plt.show()

###########################################################################

n_groups = 2
men = (lista_1M, lista_1M_s)
women = (lista_1W, lista_1W_s)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.30

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = plt.bar(index + bar_width, men, bar_width,
                 alpha=opacity,
                 color='b',
                 error_kw=error_config,
                 label='male')

rects2 = plt.bar(index, women, bar_width,
                 alpha=opacity,
                 color='r',
                 error_kw=error_config,
                 label='female')

ax.set_xlim(-bar_width,len(index)+bar_width)
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival rate 1st class')
plt.xticks(index + bar_width, ('No', 'Yes'))
plt.legend()

plt.tight_layout()
plt.show()
```
